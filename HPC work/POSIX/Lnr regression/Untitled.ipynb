```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('results_graph.csv')
data1.head()
```

```python
plt.scatter(data1.x, data1.y)
x = np.linspace(0,100,100)
y1 = x + 40
y2 = 1.2 * x + 30
y3 = 1.5 * x + 20
plt.plot(x, y1, '-r', label='y=x+40')
plt.plot(x, y2, '-y', label='y=1.2x+30')
plt.plot(x, y3, '-g', label='y=1.5x+20')
plt.legend(loc='upper left')
plt.show()
```
